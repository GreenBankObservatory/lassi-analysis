## Bump Scans 
Lassi scans 18 and 19 were a reference scan, and a signal scan consisting of the same bumps from the 2016 tests, created via a zero points file (no zernikies were used).  Examining each scan, then plotting the difference clearly shows the bumps.  These bumps are also at the expected amplitudes: the three bumps by the vertex appear close to the expected values of 300, -600 and -150 microns.

All these scans are in scan pairs: reference vs. signal.  The reference scan was meant to have no commanded actuator inputs *except* the zero points file.  However, the FITS file says that FEM was enabled - this may be why the indicated position plots (which should be minuse the zero points) is NOT flat as we would expect.  Also note that we aren't flitering out disabled actuators.

```python
# Active Surface Scan 12 correspondes to LASSI scan 18, and should be the reference scan:
# no deformations in the dish were intended, but note that FEM is enabled!
fitsName = "2019_03_27_16:31:02.fits"
fn12 = "/home/gbtdata/TINT/ActiveSurfaceMgr/%s" % fitsName
h, r, x, y, ind12, indPhi12, abs12 = plotFile(fn12, filterDisabled=False)
```

```python
# 
fitsName = "2019_03_27_16:40:48.fits"
fn13 = "/home/gbtdata/TINT/ActiveSurfaceMgr/%s" % fitsName
h, r, x, y, ind13, indPhi13, abs13 = plotFile(fn13, filterDisabled=False)
```

```python
# since the bump scan was created using zero points, we can't use the indicated values,
# but must use the absolute values, which include these zero points.  So, we plot
# the difference in these absolute points in both hoop-rib and x-y space:
from parabolas import scatter3dPlot
scatter3dPlot(h, r, abs12 - abs13, "hoop-rib bump scan")
scatter3dPlot(x, y, abs12 - abs13, "xy bump scan")
```

## Z4 Scans
LASSI scans 9 and 11 were a reference-signal scan pair in which we introduced a Z4 thermal zernike coefficient of 1714.  Once we get this into x-y space and compensate for the phi factor, once can easily see the double-saddle shape one would expect for Z4, and the peaks and troughs seem to have an amplitude equivalent to what we saw with the bump scans: 500 microns.

```python
fn3 = "/home/gbtdata/TINT/ActiveSurfaceMgr/2019_03_27_15:46:00.fits"
h, r, x, y, ind3, indPhi3, abs3 = plotFile(fn3, filterDisabled=False)
```

```python
fn5 = "/home/gbtdata/TINT/ActiveSurfaceMgr/2019_03_27_15:55:28.fits"
_,_,_,_,ind5, indPhi5, abs5 = plotFile(fn5, filterDisabled=False)
```

```python
# here we use the indicated position, compensating for the phi factor
ind35 = indPhi3  - indPhi5
scatter3dPlot(h, r, ind35, "hoop-rib Z4 scan")
scatter3dPlot(x, y, ind35, "x-y Z4 scan")
```
